### ASCII Conversion Functions
Throughout this decryption process we will be working with ASCII codes. To easily convert between strings and characters we will use two functions. The first function is `str_to_ascii(s)` which takes in characters in strings and turns them into ASCII codes using the `ord()` function. The second function is  `ascii_to_str(ascii_codes)`, which takes in ASCII codes and turns them back into characters using the `chr()` function.

String to Ascii: Takes in a string and returns a list of Ascii codes using the `ord()` function:

```python
def str_to_ascii(s):             # input of a string of characters
    return [ord(c) for c in s]   # takes every character in the input and converts it into an ascii code
```

Ascii to String: Takes in Ascii code and turns characters using the `chr()` function.

```python
def ascii_to_str(ascii_codes):                     # input of a string of ascii codes
    return ''.join([chr(n) for n in ascii_codes])  # takes every ascii code in the put and converts it to a character 
```

### Repeating the Secret Key

Due to the method of encryption, our key must match the length of our message. Because of this, we need to take into consideration that our key may be shorter than the encrypted message. In this instance, our key has to be repeated until it matches the length of our message. To ensure the key matches the length of our message, we built the function `get_padded_key_ascii(key_ascii, length` which takes inputs of a tentative `key_ascii` and a desired `length` of our message. This function uses modular airthmetic on the index of `key_ascii`, dividing by the length of `key_ascii` to keep looping through `key_ascii` until the padded key matches the length of our message.

```python
def get_padded_key_ascii(key_ascii, length): # defining function taking inputs of key_ascii and desired length 
    padded_key_ascii = []                    # created an empty list to hold incoming padded key
    
    for i in range(length):                  # sorting through every element of desired length 

                                             # preforming modular arithmetic on the index of key ascii
        padded_key_ascii.append(key_ascii[i % len(key_ascii)])

    return padded_key_ascii                  # returns the padded key 
```

### Encryption and Decryption Process

Although our goal is to decrypt a message, I am inlcuding the function `encrypt(message_ascii, key_ascii)`. It takes inputs of ascii values of a message and ascii values of a key. This function then displays the encryption process described in the project details. It repeats the key to match the length of our message, then uses $c = (m + k) \pmod {128}$ where every $m_i$ is a sequenced integer representation of our message, every $k_i$ is a sequenced integer representation of our padded key, and $c_i$ is the remainder after modulular divison by 128. The sequenced integer representation of $(c_1, c_2, ..., c_s)$ is our encrypted message.

```python
def encrypt(message_ascii, key_ascii):

    encrypted_message = [] # this is an empty set to sotre the encrypted ascii codes

                           # calls our function to repeat the key until it matches the length of the message
    padded_key_ascii = get_padded_key_ascii(key_ascii, len(message_ascii))

                           # iteratres through each message ascii value with the corresponding key ascii value
    for m,k in zip(message_ascii, padded_key_ascii):

        c = ((m + k) % 128) # applies the encryption formula from the project details page

                           # adds the encrypted value c_i to the empty list previously mentiond 
        encrypted_message.append(c)

                           # returns the full encrypted message 
    return encrypted_message
```

To decrypt our message, we will use the function `decrypt(encrypted_asii, key_ascii)` which takes input of the ASCII representation of our encrypted message as `encrypted_ascii` and a tentative key used to decrypt our message `key_ascii`. The decryption process it the opposite of the encryption process. For each number $c_i$, the integer representation of our encrypted message, we compute the remainder from the division of $c_i - k_i$ by 128. This remainder is equal to $m_i$, the integer representation of our decrypted message. That is, it returns the original ASCII values of the encrypted message.

```python
def decrypt(encrypted_ascii, key_ascii):
    
    decrypted_message = [] # creates an empty list to store the decrypted ascii codes

                           # repeats the key unti it matches the length of the encrypted message
    padded_key_ascii = get_padded_key_ascii(key_ascii, len(encrypted_ascii))

                           # iteratres through each encrypted ascii value with the corresponding key ascii value
    for c,k in zip(encrypted_ascii, padded_key_ascii):

        m = ((c - k) % 128) # applies the reverse of the encrypted formula 

                            # adds the decrypted ascii value to empty list previously defined
        decrypted_message.append(m)
        
    return decrypted_message # returns the encrypted message as a list of ascii values
```

### Cleaning Up Words for Scoring
During the process of decrypting our message, we want to remove any possible punctuation charcaters that can make actual english words fail the upcoming dictionary check. We created the function `clean_word(word)` which takes an input of some word as `word`, and converts each word to lowercase and removes punctuation.

```python
# these are the punctuations and characters we want to remove
chars_to_remove = ',.!?@#$%^&*()_-+=;:"' + "'"

def clean_word(word):
    word = word.lower()                # converts the word to lowercause to capitalization is negligable 
    
    for char in chars_to_remove:       # iterates through every punctuation and character in the list of ones we want to remove
        word = word.replace(char, '')  # removes each punctuation and character from every input of a word
        
    return word                        # returns the cleaned word
```

### Testing and Scoring all Possible Keys
Building a function tto break the encryption, we tested every from possible key from our set `possible_keys`. For each possible key we then attempt to decrypt our message and convert the resulting ASCII values back into readable text. We then split the text into individual words and count how many *cleaned* words appear in the dictionary. We then choose the possible key with the highest score as our best-matching key, saved as `found_key`.

```python
def breaking_code(encrypted_ascii, possible_keys): # takes in our list of strings of integers and list of words in dictionary

    dictionary_words = {}                          # created a dictionary to hold the score of every possible key

                                                   # this is the set of all possible words from our dictionary file
    possible_keys_set = {word for word in possible_keys}

    for possible_key in possible_keys:             # iterating through every possble key/word from our imported dictionary file

                                                   # we make that key/word and ascii code
        possible_key_ascii = str_to_ascii(possible_key)

                                                   # decrypt the message w the possible key 
        effort_ascii = decrypt(encrypted_ascii, possible_key_ascii)
        
        effort = ascii_to_str(effort_ascii)        # converting the decrypted ascii code back into readable text with words
                                                
        decrypted_words = effort.split()           # we split this into string 'words' that appear in decrypted message 

        count = 0                                  # initializing a score of 0 to quantify how 'good' each 'word' is 
        
        for decrypted_word in decrypted_words:     # iterating through each 'word' in our decrypted message

                                                   # setting each one of these words through our cleaning word function
            cleaned_word = clean_word(decrypted_word)
            
            if cleaned_word in possible_keys_set:  # if the cleaned word is in the dictionary
                count += 1                         # increase the score by 1 for each word from the key 
                
        dictionary_words[possible_key] = count     # saves the possible key and its score
        
        # we want to find the word that has the maximum value (see week 12)
        # this looks through all the words in dictionary_words and finds the word whose value is the largest
    highest_scored_key = max(dictionary_words.values())

    for key in dictionary_words:                  # finds the key that produced the highest score 

                                                  # if the key is the same of the highest scored key 
        if dictionary_words[key] == highest_scored_key:
            found_key = key                       # save the highest scored key as found_key 
            break                                 # once we found the highest scoring key, stop 
    
    return found_key                              # returns the best key
```

### Finding the Best Key
Since our `breaking_code(encrypted_ascii, posisble_keys)` function returns the possible key with the best score, we call our function store the best key as `found_key` found by testing all words from our dictionary. We can then print that stored key and can print it to see what it is.

```python
found_key = breaking_code(encrypted_ascii, possible_keys)
print(found_key)
```

### Final Decryption
Now that we've found our key with the highest score, we can do our final decryption to see our assigned encrypted message. The best key what was found was '**wonderland**'. We then turned it into an ASCII code using our function `str_to_ascii`, and save that as `found_key_ascii`. Then we use this final ASCII key in our function `decrypt` and save this output of an ASCII code as `final_ascii`. We can now use our `ascii_to_str` function to save this output as `final_message` to convert these ASCII codes back to readable text. Finally, we are able to display our encrypted message

```python
found_key_ascii = str_to_ascii(found_key)

final_ascii = decrypt(encrypted_ascii, found_key_ascii)
final_message = ascii_to_str(final_ascii)

print(final_message)
```
