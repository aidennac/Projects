## Introduction: 
In this report, we are going to explore the properties of prime numbers, leading us to discover *prime-like numbers* and *false prime numbers* through the concept of *congruency* and the use of *Fermat's Little Theorem*. To move forward, we must establish the terms:

- **Prime Number(s)**: A prime number $n$ is a positve integer $n$ > 1 such that it can only be divided by 1 and by $n$ itself. We often refer to *prime numbers* as *primes*.
  - Examples:
    - 13 is a prime number since it can only be divided by 1 and by 13 itself.
    - 12 is not a prime number since it can be divided by 1,2,3,4,6, and 12.
- **Composit Number(s)**: A composite number $n$ is a positive integer > 1 that has more than two factors. We often refer to *composite numbers* as *non primes*.
  - Examples:
     - 8 is a composite number since it has factors 1,2,4, and 8.
     - 7 is not a composite number since it can only be divided by 1 and by 7 itself.

We can now introduce the concept of *congruency* and the use of *Fermat's Little Theorem* to define a *prime-like number* and a *false prime number*:
- **Congruency**: Given arbitrary integers $i, j, k$ we can say that $i$ and $j$ are *congruent modulo (k)* if $i$ and $j$ have the same remainder when divided by $k$. We are able to denote this relationship as $i \equiv {j} \pmod {k}$.
  - Example:
>$$67 \equiv 40\pmod 9$$ which tells us 67 and 40 are *congruent modulo 9* since their remainders are both 4.
- **Fermat's Little Theorem**: If ($p$) is a prime number, then $a^p \equiv a\pmod p$ for all $0 \leq a < p$. 
  - Example: for the prime number ($p = 5$) we have the following:
>$$0^5 = 0 \equiv 0 \pmod 5$$ $$1^5 = 1 \equiv 1 \pmod 5$$ $$2^5 = 32 \equiv 2 \pmod 5$$ $$3^5 = 243 \equiv 3 \pmod 5$$ $$4^5 = 1024 \equiv 4 \pmod 5$$ which holds for values $a = 0, 1,2,3,4.$
>
- **Prime-Like Number(s)**: A positive number $n > 1$ (not necessarily a prime number) such that $ a^n \equiv a \pmod n$ for all $ 0 ≤ a < n$ is called a *prime - like number*. In the example above, we used the prime number $p = 5$ which satisfied the theorem, making $p = 5$ a **prime-like number**.

You may be wondering, "does this theorem hold for more than just prime numbers?" It turns out there are composite numbers that are prime-like as well. These numbers are called **false prime numbers**.
-  **False Prime Number(s)**: A false prime number `n` is a *composite prime-like number*. 

**Throughout this report we will use these terms as building blocks for our functions to come.**

### Finding Prime Numbers
In order to continue on our journey of finding prime-like and false prime numbers, we first need to find *prime numbers*. Using Python, we have made two functions: `is_prime(n)` and `get_primes(n)`. These help us identify and list/store *prime numbers*. Later on in this report, we will reference these functions in our journey of exploring false prime numbers.


#### `is_prime(n)`
Our first function is `is_prime(n)` which takes in integer `n` to see if it `n` is a prime number or not.

By our definition of a *prime number* $n$, we can check every integer from 2 to $n$ itself to see if there exists an integer $d$ that divides $n$ evely, indicating $n$ is *not prime* returning `False` in such case, or returning `True` if there is no such divisor $d$, indicating $n$ *is prime*. Mathematically speaking, we are checking the expression ${n} \equiv 0 \pmod d$. The downside to this operation as code is that checking every integer from 2 up to $n$ increases runtime, so this leads us to look for optimizations.

For said optimization, let us consider the case that $n$ is a composite number. That is, suppose $a,b$ are factors of $n$ such that $a \cdot b = n$ and $2 ≤ a, b ≤ n$. It holds that either $a$ or $b$ $\leq \sqrt n$. This can be shown using proof by contradiction: 
>Assume that  it is  not  the  case:  $a$ or  $b \leq \sqrt n.$  <br>
>That  is,  suppose  ${a} > \sqrt n$  and $b > \sqrt n$. <br>
> Then,  $a \cdot b > \sqrt n \cdot \sqrt n = n$. <br>
>Since we stated $a \cdot b = n$, it follows $a \cdot b \ngtr n$.<br>
>Hence, a contradiction.
>
 Due to this, we only have to check for divisors $d$ up to $\sqrt n$.

Finally, I must mention that since our `is_prime` function is utilizing python's `range` function to search for such divisor $d$, we must add a distinct `+1` to the end of our range to combat python's 0 index properties to ensure that $\sqrt n$ is iuncluded in our range.

```python
# optimized - if n is a composite, then it must have a divisor d ≤ sqrt n. 
# So we can test divisbility only up to sqrt n, which reduces the amount of check from n to sqrt n.

from math import sqrt                    # allows us to use the sqrt function
def is_prime(n):                         # function to check if n is a prime
    for d in range (2,int(sqrt(n))+1):   # this is checking the optimized divisors d ≤ sqrt n rather than n itself
        if n % d == 0:                   # if d divides n, then n is not prime
            return False                 # returns value of false for given n 
    return True                          # no divisors found so it returns true because n is prime
```

We can test our `is_prime` function by using previously stated composite and prime numbers:

```python
is_prime(8)
```

```python
is_prime(13)
```

#### `get_primes(n)`
Our second function is `get_primes(n)` which allows us to list a number of primes less than or equal to its corresponding integer input, `n`. This function builds an empty set `primes` for all incoming prime numbers, and then passes a variable `m` (starting at $m = 2$) through our aforementioned function `is_prime(n)` for as long as the variable $m \leq n$. If $m$ happens to be a prime number, we add it to our set `primes`. We then reassign $m$ by adding a distinct `+1` to it before the end of the loop to ensure that this function runs until the condition $m \leq n$ is met.

```python
def get_primes(n):               # function with variable input n to get a list of prime numbers totatling n amount.
    primes = []                  # empty set which prime numbers will reside
    m = 2
    while m <= n:                # if i were to use len(prime) it would keep going until i have n primes. 
                                 # however, we want all primes ≤ n so i want to check every number m up to n. 
        if is_prime(m):          # puts variable m through our function to see if m is prime 
            primes.append(m)     # adds prime number m to our once empty set of primes
        m += 1                   # adds 1 to m to check for the next iteration of the loop as long as m is ≤ n 
    return primes
```

To test our `get_primes` function, we will attempt to get all primes less than 20.

```python
get_primes(20)
print('Function Success! The prime numbers up to 20 are {}!'.format(get_primes(20)))
```

### Finding Prime-Like and False Prime Numbers
To find *prime-like numbers* and *false prime numbers*, we have created two functions: `is_prime_like(n)` and `is_false_prime(n)`. The `is_prime_like(n)` function is built upon our `is_prime(n)` function, as well as our solidified understanding of *congruency* and *Fermat's Little Theorem*. The instruments that make up our `is_false_prime(n)` fuction are the same as that of our `is_prime_like(n)` function including the function `is_prime_like(n)` itself. Through the remainder of this report, we will see how the functions work together and what we can do with the our outcome of false primes.

#### `is_prime_like(n)`
We head into our third function, `is_prime_like(n)` which takes in an integer `n` and returns `True` if that integer $n$ is prime-like, or `False` if $n$ is not. 

For all inputs $n$, by our definition of a prime-like number $n$, if $n$ $< 2$ then our function `is_prime_like(n)` returns `False`. If $n$ $> 2$ we check every integer, $a$, from  $2$ to $n$ to see if $a^n \pmod n = a$, giving us the exact remainder of $a$. It then compares $a^n \pmod n ≠ a$ and returns `False` for some input $n$ shall `≠` be the case. Should an input $n$ return `True`, congruence holds for all $a$. That is, $n$ is a *prime-like number*.

As shown above, our code first needs to compute the power, $a^n$. Then it preforms the modular division, $a^n \pmod n$, and lastly compares that result to see if it is $= a$ or not. That is, we are computing $a^n \pmod n ≠ a$, or $a^n \pmod n = a$. We can state the relationship that as $n$ grows, $a^n$ grows with it, in turn creating more computations, thus increasing the runtime of our code. This is clearly inefficient. To combat this, we are going to utilize pythons `pow()` function, which uses modular arithmic to compute the power and the remainder at the same time. The result of `pow(a,n,n)` is the same as $a^n \pmod n$, however it executes faster. This allows us to cut down on runtime and further optimize our code. 

Visually, we can see the inclusion of this optimization in our code by the lines that contain:
>`if pow(a,n,n) != a:`
>
>`return False`
>

```python
# i alreayd have: is_prime(n) which tells me if a number is prime and get_primes(n) which gives me all the primes equal to or less than n
def is_prime_like(n):

    if n < 2:               # number has to be greater than 2 
        return False        # if the number is less than 2, return False

    for a in range(2,n):    # tells me for all 0 ≤ a < n
        if pow(a,n,n) != a: # this computes a^n mod n and checks if it is equal to a by checking if it is false.
            return False    # if it is not equal to a then congruence is false and tells us n is not a prime-like number when one n is found
    return True             # if it finishes without failure then congruence holds for all a so n is a prime-like number
```

#### `is_false_prime(n)`
Let's us recall that *Fermat's Little Theorem* is for all prime numbers, that is, it turns out every prime is a prime-like number. However, not all prime-like numbers are prime themself. There exists positive composite prime-like numbers , which we call *false prime numbers.*

Our fourth function `is_false_prime(n)` utilizes our previous `is_prime_like(n)` function to determine if an input variable `n` is a *prime-like number* or a *false prime number*. The function `is_false_primes(n)` has to have two conditions be met in order for it to return an output value. These conditions are:
> `is_prime(n) == False)`, and <br>
> `is_prime_like(n)`. <br>
That is, $n$ is not a prime number and it is a prime-like number.

And thats it! We've defined that a false prime number is an integer $n$ such that $n$ is not prime along with $n$ is a prime-like number. We were able to use our functions `is_prime(n)` which told us if an input integer $n$ was or was not prime, and `is_prime_like(n)` which told us if n was a prime-like number or not.

To test our `is_false_prime(n)` function, let us attempt to obtain the first 20 false prime numbers: <br>
We start off by further defining our function `is_false_prime(n)` to include the necessary conditions.

```python
# 'once you know that a number is prime-like you need to just check that it is not a prime number'
# optimized: for `or` if the first part is true then stop. for `and` if the first part is true then stop

def is_false_prime(n):                                   # function to tell us if the prime-like number is prime-like but not a prime number
    return ((is_prime(n) == False) and is_prime_like(n)) # the `and` here lets us compute is_prime_like only if n is not prime
```

We then create an empty list `false_primes = []`for incoming false primes. As seen in our `get_primes(n)` function, we are using a `while` loop for the condition such that we search the length (`len`) of false primes until we have 20 false primes. Starting at the value of `n = 2`, for every pass through the loop, we check if $n$ at that current pass is a false prime using our `is_false_prime(n)` function. If it turns out that $n$ is prime, then we add such $n$ to our set `false_primes`. We then add a distinct `+ 1` to $n$ before the end of the loop to ensure that we satisfy `len(false_primes) < 20:`.

```python
false_primes = []                # this is the empty set where we will collect then state false primes
n = 2                            # starting at the first prime 2 we go up until we hit 20 false primes.
while len(false_primes) < 20:    # tells the code to only run when the length of false primes is under 20. Once 20 is it hit it stops.
    if is_false_prime(n):        # if the number n is a false prime, it will be added to our empty set false primes.
        false_primes.append(n)
    n += 1                       # the addition assignment operator adds 1 to n then resigns the new value to n for the next iteration 
                                 # of the loop until we hit 20
```

We can now print our first 20 false prime numbers!

```python
print('The first 20 false primes are sequentially {}'.format(false_primes)) # prints the first 20 false primes
```

### Prime Decomposition
In the midst of our journey of discovering and analyzing false prime numbers, we may want to take things a step further and wonder about their prime decompositions. Any integer $n > 1$ can be written as a product of prime numbers:
> $n = p_1 \cdot p_2 \cdot ... \cdot p_m$ where $p_1, ..., p_m$ are prime numbers.

This expression is know as the prime decomposition (or prime factorization) of $n$.
#### `get_prime_decomp(n)` 

Using all of our resources in form of code, we were able to create one last function `get_prime_decomp(n)`. This function breaks down any positive integer input `n` into its prime decomposition.

After defining the function `get_prime_decomp(n)`, we create an empty set `factors = []`, which will hold the incoming prime decomposition integers of the input `n`. We then call back our `get_primes(n)` function to generate a list of all prime numbers $p$ ≤ input $n$. It follows that these prime numbers provide the possible primes that are to be used in the prime decomposition of $n$. 


Then, `for` every prime decomped number $p$ in our set `primes`, we apply the use of a`while` loop to check whether the input $n$ is divisible by $p$. In such case, we append $p$ to our list `factors` and divide $n$ by $p$. This function continues until $n$ is no longer divisible py $p$. Once $n = 1$, the factorization of $n$ is complete and the function retunrs to our filled list, `factors`.


```python
def get_prime_decomp(n):
    factors = []
    primes = get_primes(n)

    for p in primes: # continues to factor until n is reduced until 1
        while n % p == 0: # as long as p div n
            factors.append(p)
            n = n // p
        if n == 1:
            break
    return factors
```
