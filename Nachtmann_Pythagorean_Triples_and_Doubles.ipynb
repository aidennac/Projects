## Introduction
**In this** report, we are going to explore elements and properties of one of the most influential formulas in mathematics, the Pythagorean Theorem. Namely, these elements are Pythagorean Doubles, Pythagorean Triples, as well as a certain property that makes these elements something we call 'primitive'. First, let us review the Pythagorean Theorem.

- **The Pythagorean Theorem**: We say that some positive integers $a,b,c$ are lengths of sides of a right triangle if and only if $a^2 + b^2 = c^2$.

**Now that** we have this fresh in our memory, we can start to wonder about the distinct triples of integers  satify the equation. Such triad of integers ($a,b,c$), are what we call *Pythagorean Triples*. To note, we define:
- **Pythagorean Triple**: A triple of positive integers $(a,b,c)$ such that $a^2 + b^2 = c^2$.
  - Example: Consider two sets of triples, ($3,4,5$) and ($4,5,6$), and let us see if either is a Pythagorean Triple.
> $$(3,4,5)   \Longrightarrow  (3)^2 + (4)^2 = (5)^2   \Longrightarrow   9 + 16 = 25 \Longrightarrow 25 = 25 $$ $$(4,5,6)   \Longrightarrow   (4)^2 + (5)^2 = (6)^2   \Longrightarrow   16 + 25 = 36 \Longrightarrow 41 ≠ 36$$
> The first triple ($3,4,5$) gives us a true equation, hence, it is a Pythagorean Triple.  
> The second triple ($4,5,6$) gives us a false equation, therefore, it is not a Pythagorean Triple.

**In our** method of discovering Pythagorean Triples ($a,b,c$) via Pythagoreans Theorem ($a^2 + b^2 = c^2$), we can see that our element $c$ is determined by the pair ($a,b$). Therefore, it is sufficient enough to find a set of doubles ($a,b$) such that we can find an element $c$ that satisfies Pythagoreans Theorem. In such case, we can call these doubles *Pythagorean Doubles*. To note, we define:
- **Pythagorean Doubles**: A set of two positive integers ($a,b$) such that $\sqrt{a^2 + b^2} = c$ where $c$ is a positive integer.
> Note: The resulting positive integer $c$ from $(a,b)$ provides us with enough elements to consider such group of numbers a Pythagorean Triple.

### Checking for Perfect Squares
In order to find Pythagorean Doubles and Pythagorean Triples through computational means, we need to be able to determine if any given integer is a perfect square. In particular, we need a way to check if the expression ($a^2 + b^2$) is a perfect square. If it is, then there exists an integer $c$ such that $a^2 + b^2 = c^2$ where the elements $(a,b,c)$ form a Pythagorean Triple.

We achieved this by creating the function `is_square(n)`. This function takes a postive integer input $n$, computes its square root, and checks whether is it equal to the square of its *rounded* square root. That is, we are checking $n = (\sqrt{n})^2$. If this equation holds, the function returns `True`, meaning $n$ a perfect square. Otherwise the function returns `False`, meaning $n$ is not a perfect square.

```python
def is_square(n):               # definition of our function
    sqrt_n = sqrt(n)            # creates a variable for the square root of our input n
    if n == round(sqrt_n)**2:   # checks condition for input n
        return True             # fnc returns true indicating a n is a perfect square
    return False                # fnc returns false indicating n is not a perfect square
```

To see if our function `is_square(n)` works, let us find all perfect squares from the integers 0 to 100.

```python
for n in range(101): 
    if is_square(n):
        print(n, 'is square')
```

As we can see, our function `is_square(n)` for finding perfect squares works. We now move on to finding Pythagorean Triples.

### Generating Pythagorean Triples

Now that we have a way to check for perfect squares, we can generate Pythagorean Triples. To do this, we created the function `get_ptripels(n)` which takes in an input $n$, and produces all triples from the pair ($a,b$) where $1 \leq a \leq n$ and $1 \leq b \leq n$. This function contains an empty list `ptrip` which will hold all incoming Pythagorean Triples. Then for each pair $(a,b)$, the function computes $(a^2 + b^2)$  and assigns the value to a variable `c_sq`. It uses our earlier function `is_square(n)`, computing `is_square(c_sq)`. This is to determine if `c_sq` is a perfect square. 
> Note: By the Pythaogrean Theorem, see that $ c^2 = a^2 + b^2 \Longrightarrow c = \sqrt{a^2 + b^2}$ <br>

In the case that it is, we take the square root of `c_sq` and assign the value to $c$. We then add the triple ($a,b,c$) to the list`ptrip`. After we have run through all of our pairs $(a,b)$, the function returns to the list `ptrip`, which contains all found Pythagorean Triples.

```python
def get_ptriples(n):                   # defining fnc w input n 
    ptrip = []                         # empty set where the pythagorean triples will live
    for a in range (1, n+1):           # first for loop needed for element a in py's thm
        for b in range (1, n+1):       # second and nested for loop needed for element b in py's thm
            c_sq = (a**2 + b**2)       # definition of c^2 through elements a and b in py's thm
            if is_square(c_sq):        # checks if this new variable is perfect square
                c = int(sqrt(c_sq))    # if it is a perf square, then we assign the integer value of the square root of that number to element c 
                ptrip.append((a,b,c))  # adds our pythagorean triple to its dedicated list
    return ptrip                       # returns to filled list of Pythagorean Triples
```

To test if our `get_ptriples(n)` function works, let us find all Pythagorean Triples for our input $n = 20$ such that $1 \leq a$ and $b \leq n$.

```python
get_ptriples(20)
```

We have now generated a list of Pythagorean Triples satisfying the Pythagorean Theorem. Now, we turn our attention to Pythagorean Doubles, which are such pairs of elements ($a,b$) that produce a Pythagorean Triple.

### Generating Pythagorean Doubles

As previously stated, it is sufficient enough to find the pair of two positive integers ($a,b$) such that the Pythagorean Theorem is satisfied. These pairs of elements are Pythagorean Doubles.

Since every Pythagorean Triple ($a,b,c$) contains the pair ($a,b$), we can generate Pythagorean Doubles directly from the output of our `get_ptriples(n)` function. It takes an input $n$, calls our `get_ptriples(n)` function, and assigns the output of Pythagorean Triples for $n$ to a variable called `pdub` (short for Pythagorean Doubles). It then uses Python's 0-indexing properties to extract the first two elements from the provided triples. Namely, the first element of each triple, $a$, is obtained by `t[0]` and the second element of each triple, $b$, is obtained by `t[1]`. Our function then returns these pairs from each triple found.
> Note: It is worth mentioning that this function preserves the order of pairs, so if both $(a,b,c)$ and $(b,a,c)$ are triples that we encounter, then both $(a,b)$ and $(b,a)$ will appear in the the output.

```python
def get_pdoubles(n):                          # defining function with input n
    pdub = get_ptriples(n)                    # takes output of triples
    return [ (t[0], t[1]) for t in pdub]      # returns list of selected elements to build our doubles
```

To test if our `get_pdoubles(n)` function works, let us find all Pythagorean Doubles for input $n = 20$.

```python
get_pdoubles(20)
```

As we can see, these Pythagorean Doubles exactly match the first two elements of each Pythagorean Triple when input $n = 20$. We may now pose the question, what do these Pythagorean Doubles look like? To answer, let us consider plotting these doubles.

### Primitive Pythagorean Doubles and Primitive Pythagorean Triples
Motivated by this distinction, we turn our attention to those doubles and triples which are primitive. We introduce the idea of *Primitive Pythagorean Triples* and *Primitive Pythagorean Doubles*. We define:
- **Primitive Pythagorean Triples**: A Pythagorean Triple ($a,b,c$) is *primitive* if the greatest common factor of the integers $a,b,c$, is 1.
- **Primitive Pythagorean Doubles**: A Pythagorean Double ($a,b$) is *primitive* if the greatest common factor of the pair ($a,b$) is 1.
> These definitions tell us that, in order to be *primitive*, a Pythagorean Triple ($a,b,c$) or Pythagorean Double $(a,b)$ cannot be written as a multiple of a smaller triple or double.

Now that we have definied what makes Pythagorean Doubles and Triples *primitive*, we can construct a function that computes the greatest common factor so that we can determine the Primitive Pythagorean Doubles and Triples.

### Using Euclidean's Algorithm

One way to compute the greatest common factor of a pair of integers ($a,b$) is by using Euclidean's Algorithm. This algorithm is built upon the following facts about a common divisor $d$:
> 1.  If $d$ divides both $a$ and $b$, then $d$ must also divide $b - a$
> 2.  If $d$ divides both $a$ and $b - a$, then $d$ must also divide $b$.
> 
It follows that the greatest common factor of $a$ and $b$ is the same as the greatest common factor of $a$ and $b - a$. Repeating the subtraction in these condictions leads to the remainder form of Euclidean's Algorithm, which is how we will apply the use of modular division in our code. 

After introducing the underlying conditions of Euclidean's Algorithm, we constructed the function `Euclidean_algorithm(a,b)` to compute the greatest common factor of integer pairs which appear in Pythagorean Doubles and Pythagorean Triples. This function takes an ordered pair ($a,b$) as input, and replaces the pair's elements with their absolute values. It ensures that $a \leq b$ by swapping the values whenever $a > b$. It then computes the remainder $b \pmod a$ and replaces $b$ with that remainder. The function runs until $b = 0$, which is the point where the remaining value $a$ is the greatest common factor of the original pair $(a,b)$.

```python
def Euclidean_algorithm(a,b):      # defining fcn to compute gcf of integers a and b 
    a = abs(a)                     # makes a non-negative
    b = abs(b)                     # makes b non-negative 
    while (a > 0) and (b > 0):     # continues until either a or b is 0
        if a > b:                  # helps keep the smaller value in the pair
            a,b = b,a              # swaps and b if needed
        b = b % a                  # replaces b with the remainder when b is divided by a 
    return a                       # returns the gcd of a 
```

To test our function `Euclidean_algorithm(a,b)`, let us find the greatest common factors for some arbitrary Pythagorean Doubles.

```python
print('The Pythagorean Double (10,15) has the greatest common factor {}.'.format(Euclidean_algorithm(10,15)))
print('The Pythagorean Double (9,12) has the greatest common factor {}.'.format(Euclidean_algorithm(9,12)))
print('The Pythagorean Double (99,144) has the greatest common factor {}.'.format(Euclidean_algorithm(99,144)))
```

As we can see, our function `Euclidean_algorithm(a,b)` correctly computes the greatest common factor for each given Pythagorean Double. The greatest common factors of (10,15), (9,12), and (99,144) are 5, 3, and 9, respectively. Since each result is greater than 1, this tells us that these doubles are not primitve. Thus, computing the greatest common factors allows us to indentify Primitive Pythagorean Doubles, which we will now use to generate Primitve Pythagorean Triples and Primitve Pythagorean Doubles.

### Generating Primitive Pythagorean Triples
Now that we have defined what properties and algorithms make a Pythagorean Triple Primitive, we can put all of our pieces together into one final assembly. We created the function `primitive_ptriples(n)` which takes in an input $n$, calls our prior function  `get_ptriples(n)` and assigns the resulting list of Pythagorean Triples to the variable `triples`. Using Python's indexing, it then checks each triple, $t = (a,b,c)$ by appling`Euclidean_algorithm(t[0],t[1])` to the pair $(a,b)$. If the greatest common factor of $a$ and $b$ is 1, then the triple is primitve and is included in the output of this function.

```python
def primitive_ptriples(n):
    triples = get_ptriples(n)                                                # create a variable for our list 
    return [ t for t in triples if Euclidean_algorithm (t[0], t[1]) == 1 ]   # so t is (a,b,c) , t[0] = a ,t[1] = b, so if the alorithim 
                                                                             # says the gcd is 1 then it is primitive 
```

### Generating Primitive Pythagorean Doubles
Just as we were able to generate Pythagorean Doubles from the output of the function `get_ptriples(n)`, we can generate Primitive Pythagorean Doubles from the output of our function `primitive_ptriples(n)`. We created the function `primitive_pdoubles(n)` which calls our function `primitive_ptriples(n)` and stores the resulting Primitive Pythagorean Triples in a variable `prim_db`. We then use Python's indexing to extract the first two entries of each triple. To be precise, `t[0]` gives us $a$ and `t[1]` gives us $b$, resulting in the primitive pair, $(a,b)$.

```python
def primitive_pdoubles(n):
    prim_db = primitive_ptriples(n)        # we create a variable for our list of pythagorean triples so we can take elements (a,b) out.
    return [(t[0],t[1]) for t in prim_db]  # returns t[0] as a and t[1] as to provide the primitive pair (a,b) as output
```
